```python
import matplotlib.pyplot as plt

print("hi")
x = [1, 2, 3, 4, 5, 6]
y = [4, 2, 8, 3, 1, 4]
c = ["r", "b", "y", "g", "m", "c"]  # 6 colors for 6 bars

plt.xlabel("Languages")
plt.ylabel("No. of People")

plt.bar(x, y, color=c, label="plot", width=0.5, align="center",linestyle=":", edgecolor="b", alpha=0.9)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("hi")
x = ["Python", "C", "C++", "Java"]
y = [85, 60, 63, 82]
n = [12, 4, 5, 9]

width = 0.3
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.figure(figsize=(8, 5))  # Optional: set figure size

plt.xlabel("Programming Languages", fontsize=15)
plt.ylabel("Number of Users (in millions)", fontsize=15)
plt.title("Programming Language Popularity Comparison", fontsize=15)

plt.bar(p, y, color="r", width=width, label="2024 Users")
plt.bar(p1, n, color="y", width=width, label="2023 Users")

plt.xticks(p + width / 2, x, rotation=50)
plt.legend()

# Add value labels
for i in range(len(p)):
    plt.text(p[i], y[i] + 1, str(y[i]), ha='center', fontsize=9)
    plt.text(p1[i], n[i] + 1, str(n[i]), ha='center', fontsize=9)

plt.tight_layout()
plt.show()

```

```python
#scaterplot
import matplotlib.pyplot as plt
print("hi")
day=[1,2,3,4,5,6]
no=[2,3,4,5,6,7]
c=["r",'b',"y",'g',"r","b"]
plt.title("scaterplot",fontsize=20,color="b")
plt.xlabel("day",fontsize=20,color="g")
plt.ylabel("no",fontsize=20,color="y")
sizes=[1000,500,700,350,1100,780]
plt.scatter(day,no,color=c,s=sizes,marker="*",edgecolors="r",linewidth=3)



plt.show()
```
